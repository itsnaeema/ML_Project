```python

#in multiple linear regression- munltiple indepent variable and one dependent varibable
#here sales id dependent variable others are independent variable
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/advertising.csv")
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#seaborn()-to understand effect of each independent variable dependent variablmne ath reethiyl effect cheyyun graphycally representng using seaborn
#regplot()-regression plot
import seaborn as sns
sns.regplot(x=df['TV'],y=y,color='m')#x=tv,y=sales data
plt.title('TV vs sale')
```

```python
sns.regplot(x=df['Radio'],y=y,color='y')#x=radio,y=sales data
plt.title('Radio vs sale')
```

```python
sns.regplot(x=df['Newspaper'],y=y,color='b')#x=newspaper,y=sales data
plt.title('NEWSPAPER vs SALE')
```

```python
#training and testingdata
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[200,40,5.6]]))
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_pred-y_test})
df1
```

```python
print("slope is",model.coef_)
print(list(zip(x,model.coef_)))
```

```python
print("constant is",model.intercept_)
```

```python
#performance evaluation
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
print("error is",mean_absolute_error(y_test,y_pred))
print("percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("R2 score is",r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
data=mean_squared_error(y_test,y_pred)
print(data)
```

```python
print(np.sqrt(data))
```

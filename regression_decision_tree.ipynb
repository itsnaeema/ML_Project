```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/advertising.csv")
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
import seaborn as sns
sns.regplot(x=df['TV'],y=y,color='m')
```

```python
sns.regplot(x=df['Radio'],y=y,color='r')
```

```python
sns.regplot(x=df['Newspaper'],y=y,color='b')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_pred-y_test})
df1
```

```python
#performance evaluation
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
print("error is",mean_absolute_error(y_test,y_pred))
print("percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("R2 score is",r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
data=mean_squared_error(y_test,y_pred)
print(data)
```

```python
print(np.sqrt(data))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/melon.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y)
plt.xlabel("weight of watermelon")
plt.ylabel("price of watermelon")
plt.title("WATERMELON")
```

```python
#training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#create a scatter based on xtrain and y train
plt.scatter(x_train,y_train)
#creatin a regression line using plot by passing xtest and ypred
plt.plot(x_test,y_pred,color='m')
```

```python
print(model.predict([[12]]))
```

```python
print('Slope is',model.coef_)
```

```python
print("constant is",model.intercept_)
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_pred-y_test})
df1
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
print("error is",mean_absolute_error(y_test,y_pred))
print("percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("R2 score is",r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
data=mean_squared_error(y_test,y_pred)
print(data)
```

```python
print(np.sqrt(data))
```

```python


#Decision tree
#it is an example of supervised machine learning algorithm
#used for both regression and classifiaction method
import numpy as np
import pandas as pd
df=pd.read_csv("/content/Iris.csv")
df
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
#training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#creating model
from sklearn.tree import DecisionTreeClassifier#we can also use decision tree in regression model,so we import DecisionTreeRegressor
model=DecisionTreeClassifier(criterion='entropy')#pass entropy as argument to calculate the value of entropy
#in regression modell no need to pass criterion=entropy,also we cannot draw or plot tree
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#performance evaluation
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
#Tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))#defining figure size
tree.plot_tree(model,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],class_names=['Iris-versicolor','Iris-virginica','Iris-setosa'],filled=True)
#passing input features(node) as list and leaf in in clqass names(paass in order of y_train(output))
```

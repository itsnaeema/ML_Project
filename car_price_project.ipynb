```python
#Encoding
#get_dummies-used to overcome the drwaback of onehot encoding ,suitable if we have more objects in column
#in het dummies first coluimn will drop
df6=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df6
```

```python
#joining to dataframe oldoe and new one
#df6 have only the encoded value so we conctnate both df
#concatnation
dfc=pd.concat([df,df6],axis=1)
dfc
```

```python
#droping  unwnted column(unamed,name,Fuel_Type	,Transmission	,Owner_Type,location(we encoded these columns so  we r droping)
dfc.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1,inplace=True)
dfc
```

```python
dfc.dtypes#here mileage ,power,and engine are object because of unit ,so we want to split the string ,to change onlyu into numerical
```

```python
#split
dfc['Mileage']=dfc['Mileage'].str.replace('km/kg','')
dfc['Mileage']=dfc['Mileage'].str.replace('null','0')
dfc
```

```python
dfc['Mileage']=dfc['Mileage'].str.replace('kmpl','')
dfc['Mileage']=dfc['Mileage'].str.replace('null','0')
dfc
```

```python
dfc['Engine']=dfc['Engine'].str.replace('CC','')
dfc['Engine']=dfc['Engine'].str.replace('null','0')
dfc
#there are sum null values in milage ,power engine in the given file,so we want replace that null value to become numerical
```

```python
dfc['Power']=dfc['Power'].str.replace('bhp','')
dfc['Power']=dfc['Power'].str.replace('null','0')
dfc
```

```python
dfc.dtypes#here again mileage,engine,power are in object to change that we should convert into float,(change the datatype)
```

```python
#data type converstion
dfc['Engine']=dfc['Engine'].astype(float)
dfc['Mileage']=dfc['Mileage'].astype(float)
dfc['Power']=dfc['Power'].astype(float)
```

```python
dfc.dtypes
```

```python
#first change the null value into 0>'0' pass as an object,so,mileage,power,engine will not be in 0 value so we should replace into numerical value,
#we cannot firstly change the value because it was in an string form no we convert in number
#Replace a numerical Value,Numerical values evide ellam zeros und avde missin value ayi treat cheythu
#here if it is numner using numpy 'np.' replace,if it is striing replace using str.replace
# replace 0 as a numrical value
# treat 0 as missing value
dfc.loc[dfc.Engine==0,'Engine']=np.NaN
dfc.loc[dfc.Mileage==0,'Mileage']=np.NaN
dfc.loc[dfc.Power==0,'Power']=np.NaN
```

```python
dfc.isna().sum()
```

```python
#filling missn value
dfc['Mileage']=dfc['Mileage'].fillna(dfc['Mileage'].mean())
dfc['Engine']=dfc['Engine'].fillna(dfc['Engine'].mean())
dfc['Power']=dfc['Power'].fillna(dfc['Power'].mean())
dfc['Seats']=dfc['Seats'].fillna(dfc['Seats'].mode()[0])
```

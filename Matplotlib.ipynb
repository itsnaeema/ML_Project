```python
#to pass title,metion content of x and y
height=[170,150,155,160,161,180,75]
weight=[70,51,53,64,65,82,70]
plt.plot(height,weight)
plt.xlabel('Height in cm',color='m',size=10)
plt.ylabel('weight in kg',color='y',size=10)
plt.title('height-weight graph',color='g',size=30)
```

```python
#velocity-displacement graph
velocity=[5,9]
displacement=[8,0]
plt.plot(velocity,displacement)
plt.xlabel('velocity',color='r')
plt.ylabel('displacment',color='y')
plt.title('Velocity-Displacement Graph',color='m',size=20)
```

```python
#scatter plot-each of these data points looks scattered around the graph
x1=np.array([1,5,6,3,9])
y1=np.array([4,6,2,6,10])
x2=[4,3,6,8,2]
y2=[6,2,10,9,1]
plt.scatter(x1,y1,color='y')
plt.scatter(x2,y2,color='m')
```

```python
x=[1,3,5,6,7]
y=[6,3,1,7,8]
plt.plot(x,y)
plt.xlabel('x in cm',size=10)
plt.ylabel('y in cm',size=10)
plt.title('x-y GRAPH',color='m',size=20)
plt.grid(color='red',linestyle='dotted',linewidth=3)#we can also pass axis='x'  or axis ='y'
#we can als change color,width,style of line
```

```python
#subplot: to polot 2 lines in different graph(sub graph)
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1)#(val1,val2,val3) val1=row val2=column val3=order
plt.plot(x1,y1)
plt.xlabel('Velocity')
plt.ylabel('Displacement')
plt.title('Velocity-Displacement Graph')
x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)#(val1,val2,val3)
plt.plot(x2,y2)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height-Weight')
plt.suptitle('Title',size=20)
```

```python
#barchart
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.bar(x1,y1)
```

```python
#pie chart
y=[10,20,30,40]#in pie chart also only need one axis #10/100 20/100 30/100 40/100
mylabel=['apple','orange','banana','mango']#adding content(name)
mycolor=['yellow','red','hotpink','green']#also can set colors
plt.pie(y,labels=mylabel,shadow=True,colors=mycolor,autopct='%1.1f%%')#autopct-(auto percentage)-to show percentage
plt.legend()#showing colors in small box
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/company_sales_data.csv')
df
```

```python
 #Read Total profit of all months and show it using a line plot
import matplotlib.pyplot as plt
month_data=df['month_number'].tolist()#TO SELECT coloumns also can use in iloc function (x axis)
Total_profit=df['total_profit'].tolist()#y axis
plt.plot(month_data,Total_profit)
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.xticks(month_data)#to change axis number
plt.yticks([100000,200000,300000,400000,500000])
plt.title(['Company profit per month'])
```

```python
# Get total profit of all months and show line plot with the following Style properties
import matplotlib.pyplot as plt
month_data=df['month_number'].tolist()#TO SELECT coloumns also can use in iloc function (x axis)
Total_profit=df['total_profit'].tolist()#y axis
plt.plot(month_data,Total_profit,marker='o',color='red',linewidth=3,linestyle='dotted')
plt.xlabel('Month Number')
plt.ylabel('Sold unit number')
plt.xticks(month_data)#to change axis number
plt.yticks([100000,200000,300000,400000,500000])
plt.title(['Company profit per month'])
```

```python
face_cream=df['facecream'].tolist()
face_wash=df['facewash'].tolist()
tooth=df['toothpaste'].tolist()
bath=df['bathingsoap'].tolist()
sham=df['shampoo'].tolist()
mos=df['moisturizer'].tolist()
plt.plot(month_data,face_cream,marker='s',color='r',linewidth=3,label='face_cream_data')
plt.plot(month_data,face_wash,marker='s',color='g',linewidth=3,label='face_wash_data')
plt.plot(month_data,tooth,marker='s',color='b',linewidth=3,label='toothpaste_data')
plt.plot(month_data,bath,marker='s',color='k',linewidth=3,label='soap_data')
plt.plot(month_data,sham,marker='s',color='m',linewidth=3,label='shampoo_data')
plt.plot(month_data,mos,marker='s',color='y',linewidth=3,label='moisturizer_data')
plt.xticks(month_data)
plt.legend(loc='upper right')
```

```python
#4.Read toothpaste sales data of each month and show it using a scatterplot
plt.scatter(month_data,tooth)
```

```python
#5.Read face cream and facewash product sales data and show it using the bar chart with grid line
plt.bar([a-0.25 for a in month_data],face_cream,width=0.25,label='face_cream ',align='edge')
plt.bar([a+0.25 for a in month_data],face_wash,width=0.25,label='face_wash ',align='edge')
plt.xticks(month_data)
plt.legend(loc='upper right')
plt.grid()
```

```python
#6.Read the total profit of each month and show it using the histogram to see the most common profit ranges
plt.hist(month_data)
```

```python
#7.Calculate total sale data for last year for each product and show it using a Pie chart
Product=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
salesdata=[sum(face_cream),sum(face_wash),sum(tooth),sum(bath),sum(sham),sum(mos)]
mycolor=['yellow','red','hotpink','green','black','brown']
plt.pie(salesdata,labels=Product,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.legend(loc='upper left')
plt.title('Sales Data')
```

```python
#8. Read Bathing soap facewash of all months and display it using the Subplot
bathing_soap=df['bathingsoap'].tolist()
face_wash=df['facewash'].tolist()
plt.subplot(1,2,1)
plt.plot(month_data,bathing_soap)
plt.xlabel('Month number')
plt.xticks(month_data)
plt.ylabel('Soap Data')
plt.title('Soap Month Data')
plt.subplot(1,2,2)
plt.plot(month_data,face_wash,color='m')
plt.xlabel('Month number')
plt.xticks(month_data)
plt.ylabel('facewash_data')
plt.title('facewash month data ')
plt.suptitle('Month data')
```

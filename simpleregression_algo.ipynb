```python
#based on year of experience predict salary in this data salary is depenendt variable and year is independant
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y)
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.title('Year-Salary Graph')
```

```python
#traing and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#in regression model no need to do normaliztion
#model creation(regression model)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#create a scatter based on xtrain and y train
plt.scatter(x_train,y_train)
#creatin a regression line using plot by passing xtest and ypred
plt.plot(x_test,y_pred,color='m')
```

```python
print(model.predict([[8.1]]))
```

```python
#y=mx+c
#m-slope
#c-constent
#Slope(coef_)
print('Slope is',model.coef_)#modelname.coef
```

```python
#constant(intercept_)
print("constant is",model.intercept_)#y=9339.08*8.2+25918
```

```python
#difference btw actual value and predicted vale
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_pred-y_test})
df1
```

```python
#performance evaluation
#1)MAE:Mean Absolute Error
   #y=actualvale,y^=predictedvalue,y-y^=error,eqn=1/n |y-y^|
#2)MSE:Mean Square Error
   #eqn=1/N (y-y^)^2
#3)MAPE:Mean Absolute percentage Error
#4)RMSE:Root Mean Squared Error
#5)R2_Score
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
print("error is",mean_absolute_error(y_test,y_pred))
print("Percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("R2_score is",r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
data=mean_squared_error(y_test,y_pred)
print(data)
```

```python
#sqrt()
print(np.sqrt(data))
```

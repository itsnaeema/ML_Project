```python
df.drop(['CustomerID','Genre'],axis=1,inplace=True)
```

```python
df.columns
```

```python
#there is only input values
x=df
x
```

```python
#Elbow points
from sklearn.cluster import KMeans
wcss=[]#[var1,var2,var3..]
for i in range(1,11):# if i=1 2 3..
  data=KMeans(n_clusters=i,init='k-means++',random_state=42)
  data.fit(x)
  wcss.append(data.inertia_)#inertia_ : it is the function used to find varience
```

```python
#finding value of k
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss,color='r')
plt.xlabel("No.of clusters")
plt.ylabel("Varience")
plt.title("Elbow points")
```

```python
from sklearn.cluster import KMeans
data1=KMeans(n_clusters=6,init='k-means++',random_state=42)
data1.fit(x)
y=data1.predict(x)
y
```

```python
#appending output to dataframe by giving name cluster
x['cluster']=y
x
```

```python
#supervised machine learning
#split x and y
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#model creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,y_pred)
cmd
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import classification_report#all preformance evalution
report=classification_report(y_test,y_pred)
print(report)
```

```python
#Percentile
#q1 : 25 percentile   q2:50 percentile  q3:75 percentile
#Outliers
```

```python
#how to find a percentile(q1,q2,q3)
#3,5,7,1,1,8,4,6
#first we want to sort the data into ascending order from the given data
#1,1,3,4,  5,6,7,8
#then find (median) =4+5/2=4.5  (dividing in 2 part given data,taking,4&5/2(median))if it is ood take middle part,or even do like this
#this is q2:50 percentile

#to calculate q1
#calculate the median of first part(1,1,3,4)
#q1:1+3/2=2 q1=25%

#to calculate q3
#calculate the median of 2nd part(5,6, 7,8)
#q3=6+7/2=6.5

#to remove outlier we should find outliers using IQR
#calculate IQR(InteQuartileRange), Eqn=q3-q1
#q3=6.5 q1=2 ,IQR=6.5-2=4.5

#using this we find outliers,there will be outliers in bottom and top
#to find outliers in Top :(1.5*iqr)+q3=(1.5*4.5)+6.5=13.25
#if any poimt in given data is above 13.25 the it is an outlier
#outliers in bottom:(q1-1.5*iqr)=(2-1.5*4.5)=-4.75
#if any point in given data is below 4.75 then it is an outliers

```

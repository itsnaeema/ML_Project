```python
df.drop('Loan_ID',axis=1,inplace=True)
```

```python
#missing value
x=df['Gender'].mode() [0]
y=df['Married'].mode() [0]
z=df['Dependents'].mode() [0]
w=df['Education'].mode() [0]
h=df['Self_Employed'].mode() [0]
l=df['LoanAmount'].mean()
v=df['Loan_Amount_Term'].mean()
b=df['Credit_History'].mean()
df['Gender'].fillna(x,inplace=True)
df['Married'].fillna(y,inplace=True)
df['Dependents'].fillna(z,inplace=True)
df['Education'].fillna(w,inplace=True)
df['Self_Employed'].fillna(h,inplace=True)
df['LoanAmount'].fillna(l,inplace=True)
df['Loan_Amount_Term'].fillna(v,inplace=True)
df['Credit_History'].fillna(b,inplace=True)
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#one Hot Encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['Gender','Married','Dependents','Education','Self_Employed','Property_Area']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
knn=KNeighborsClassifier(n_neighbors=7)
base=GaussianNB()
model=SVC()
lst=[knn,base,model]
```

```python
for i in lst:
  print("model name is...",i)
  i.fit(x_train,y_train)
  print("prediction value...")
  y_pred=i.predict(x_test)
  print(y_pred)
  print("confusion matrix is..")
  print(confusion_matrix(y_test,y_pred))
  print("Accuracy score is....")
  print(accuracy_score(y_test,y_pred))
```

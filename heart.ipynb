```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
#Training and Testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()#making an object for the functn
scaler.fit(x_train) #creating model by passn input x_train,only fit training data(x_train),bcs u and s training data
x_train=scaler.transform(x_train)#normalization
x_test=scaler.transform(x_test)#normalization
x_train
```

```python
x_test
```

```python
#model creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,y_pred)
cmd
```

```python
score=accuracy_score(y_test,y_pred)
score
```

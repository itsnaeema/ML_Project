```python
x=df.iloc[:,:-1].values#to print as array not as data frame,we use values#x always 2 dimension
x#to print all columns except last columns
#seperating input and output ,x-assigning only input values
```

```python
y=df.iloc[:,-1].values#y is always one dimension #output
y#to print only last column (seperating input and output)
```

```python
#Training and Testing Data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Normalization
#all input columns values make as same range
#StandardScaler: it is an technique for normalization
#z=(x-u)/s this equation work based on standardScaler
#z-value to find
# x-value that given in the data
#u-mean of training data
#s-stanadard deviation of training data
#preprocessing:module
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()#making an object for the functn
scaler.fit(x_train) #creating model by passn input x_train,only fit training data(x_train),bcs u and s training data
x_train=scaler.transform(x_train)#normalization
x_test=scaler.transform(x_test)#normalization
x_train
```

```python
x_test
```

```python
#model creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)#we create model by training data of input and output
y_pred=knn.predict(x_test)#x_test is the data to predict
y_pred

```

```python
y_test
```

```python
#4 performance evaluation method
#1)Acuracy score:if the accuracy score of the model is greater than 70 then it is an good model
#now only we check acurracy score
#2)Recall:
#3)Precision
#4)FI score

#confusion matrics
#there are 4 technical terms:
#diabetics>>1,nondibet>>ic0
#TP(True postive):if the person is diabetic patient and machine predict correctly the person is diabetic 1===>1 ie,(1,1)(truely postive)
#TN(True Negative):if the person is nondiabetic patient and machine predict correctly the person is  non diabetic 0===>0 ie,(0,0)(truely neg)
#FP(False postive):if the person is nondiabetic patient and machine predict false the person is diabetic 0===>1 ie,(0,1)(falsely postive)
#FN(False negative):if the person is diabetic patient and machine predict false the person is  non diabetic 1===>0 ie,(1,0)(falsely neg)
#eg:y_test=001011100100101101110 (actual value y test)
#   y_pred=011101010110100011100 (predicted value y pred)
#TP=6(1,1)
#TN=5(0,0)
#FP=5(0,1)
#FN=5(1,0)
#        TP  FP
#        FN  TN   friste diagonal is true
##all performance measurement work based on their confusion matrics
#accuracy score=TP+TN/TP+TN+FP+FN
#after we creating a model we should check the accuracy score, if it is above 70% it is an good model

# performance evaluation
from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,y_pred)
cmd


```

```python
#we use random_state=  in testing and training data for make the accuracy value constant
#randaom state value can be 1,0,42 these are the default value to give a good performnce
score=accuracy_score(y_test,y_pred)#when run the pgm again and again the accuracy score change gradually,bcs of traing(70 data) and tesitng data(30)
score
```

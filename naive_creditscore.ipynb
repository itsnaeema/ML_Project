```python
x=df['Gender'].mode() [0]
y=df['Income'].mean()
df['Gender'].fillna(x,inplace=True)
df['Income'].fillna(y,inplace=True)
df.isna().sum()
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])#encoding like this bcs these datas are passed as dataframe,in tennis the datas are in the form of list
df# and we define column of gender df(gender) like this
```

```python
df['Education']=lab.fit_transform(df['Education'])
df
```

```python
df['Marital Status']=lab.fit_transform(df['Marital Status'])
df
```

```python
df['Home Ownership']=lab.fit_transform(df['Home Ownership'])
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
#training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#normalization(other than standard scaler there is minmax scaler also)
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#model creation
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,y_pred)
cmd
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
lab=['Average','High','Low']#y_train pass inside(output)
cm=ConfusionMatrixDisplay(cmd,display_labels=lab)
cm.plot()
```
